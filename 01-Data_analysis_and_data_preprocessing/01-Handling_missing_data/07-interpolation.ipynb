Interpolation is a useful technique to estimate or predict values between known data points.

### 7.1 Data Interpolation with Pandas

Scikit-Learn itself does not offer specific functions for data interpolation.

Some common interpolation methods available in Pandas include:

1. linear: Performs linear interpolation to fill missing values based on the nearest available data points.
2. time: Works with DataFrame or Series having a DatetimeIndex to perform interpolation based on time.
3. index: Interpolates based on the index (row labels) of the DataFrame or Series.
4. nearest: Fills missing values by using the nearest available data point.
5. zero: Replaces missing values with 0.
6. slinear: Performs linear interpolation, similar to linear.
7. quadratic: Interpolates using a quadratic polynomial.
8. cubic: Performs cubic interpolation.
9. spline: Uses cubic splines to interpolate the missing values.
10. polynomial: Interpolates using a user-defined polynomial order.
11. pchip: Interpolates using Piecewise Cubic Hermite Interpolating Polynomial.
12. akima: Interpolates using the Akima method.
13. barycentric: Applies barycentric interpolation.
14. krogh: Interpolates using the Krogh method.
15. piecewise_polynomial: Uses a piecewise polynomial function to interpolate.
16. spline: Performs spline interpolation.

Pandas offers the interpolate() method, which can be applied directly to a DataFrame to handle missing values using interpolation. This method provides several interpolation techniques that can be specified using the method parameter.

Let's demonstrate how to use Pandas for data interpolation:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generate new data with 100 points following a sinus function with larger amplitude and noise
x = np.linspace(1, 10, 100)
y = 20 * np.sin(2 * np.pi * x / 10) + np.random.normal(0, 5, 100)

# Create a DataFrame with missing values
missing_indices = np.random.choice(np.arange(100), size=40, replace=False)
y[missing_indices] = None
data = {'X': x, 'Y': y}
df = pd.DataFrame(data)

# Interpolate missing values using different interpolation methods
df_linear = df.interpolate(method='linear', axis=0)
df_spline = df.interpolate(method='spline', order=2, axis=0)
df_polynomial = df.interpolate(method='polynomial', order=2, axis=0)
df_nearest = df.interpolate(method='nearest', axis=0)
df_zero = df.interpolate(method='zero', axis=0)
df_index = df.interpolate(method='index', axis=0)
df_quadratic = df.interpolate(method='quadratic', axis=0)

# Plot the original and interpolated datasets for comparison
fig, axs = plt.subplots(4, 2, figsize=(30,30))

# Original DataFrame
axs[0, 0].scatter(df['X'], df['Y'], label='Original Data', marker='o', color='blue')
axs[0, 0].set_title('Original Data')
axs[0, 0].set_xlabel('X')
axs[0, 0].set_ylabel('Y')
axs[0, 0].legend()

# Interpolated using linear interpolation
axs[0, 1].scatter(df['X'], df['Y'], label='Linear Interpolation', marker='o', color='blue')
axs[0, 1].scatter(df_linear['X'].loc[missing_indices], df_linear['Y'].loc[missing_indices], label='Interpolated', marker='x', color='red')
axs[0, 1].set_title('Linear Interpolation')
axs[0, 1].set_xlabel('X')
axs[0, 1].set_ylabel('Y')
axs[0, 1].legend()

# Interpolated using spline interpolation
axs[1, 0].scatter(df['X'], df['Y'], label='Spline Interpolation', marker='o', color='blue')
axs[1, 0].scatter(df_spline['X'].loc[missing_indices], df_spline['Y'].loc[missing_indices], label='Interpolated', marker='x', color='red')
axs[1, 0].set_title('Spline Interpolation')
axs[1, 0].set_xlabel('X')
axs[1, 0].set_ylabel('Y')
axs[1, 0].legend()

# Interpolated using polynomial interpolation
axs[1, 1].scatter(df['X'], df['Y'], label='Polynomial Interpolation', marker='o', color='blue')
axs[1, 1].scatter(df_polynomial['X'].loc[missing_indices], df_polynomial['Y'].loc[missing_indices], label='Interpolated', marker='x', color='red')
axs[1, 1].set_title('Polynomial Interpolation')
axs[1, 1].set_xlabel('X')
axs[1, 1].set_ylabel('Y')
axs[1, 1].legend()

# Interpolated using nearest interpolation
axs[2, 0].scatter(df['X'], df['Y'], label='Nearest Interpolation', marker='o', color='blue')
axs[2, 0].scatter(df_nearest['X'].loc[missing_indices], df_nearest['Y'].loc[missing_indices], label='Interpolated', marker='x', color='red')
axs[2, 0].set_title('Nearest Interpolation')
axs[2, 0].set_xlabel('X')
axs[2, 0].set_ylabel('Y')
axs[2, 0].legend()

# Interpolated using zero interpolation
axs[2, 1].scatter(df['X'], df['Y'], label='Zero Interpolation', marker='o', color='blue')
axs[2, 1].scatter(df_zero['X'].loc[missing_indices], df_zero['Y'].loc[missing_indices], label='Interpolated', marker='x', color='red')
axs[2, 1].set_title('Zero Interpolation')
axs[2, 1].set_xlabel('X')
axs[2, 1].set_ylabel('Y')
axs[2, 1].legend()

# Interpolated using index interpolation
axs[3, 0].scatter(df['X'], df['Y'], label='Index Interpolation', marker='o', color='blue')
axs[3, 0].scatter(df_index['X'].loc[missing_indices], df_index['Y'].loc[missing_indices], label='Interpolated', marker='x', color='red')
axs[3, 0].set_title('Index Interpolation')
axs[3, 0].set_xlabel('X')
axs[3, 0].set_ylabel('Y')
axs[3, 0].legend()

# Interpolated using quadratic interpolation
axs[3, 1].scatter(df['X'], df['Y'], label='Quadratic Interpolation', marker='o', color='blue')
axs[3, 1].scatter(df_quadratic['X'].loc[missing_indices], df_quadratic['Y'].loc[missing_indices], label='Interpolated', marker='x', color='red')
axs[3, 1].set_title('Quadratic Interpolation')
axs[3, 1].set_xlabel('X')
axs[3, 1].set_ylabel('Y')
axs[3, 1].legend()

plt.tight_layout()
plt.show()
```
